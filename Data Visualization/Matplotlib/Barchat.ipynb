```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Vertical & Horizontal BarChart

```python
subjects = ['Physics', 'Maths', 'Computer Science', 'Chemistry']
students = [35, 55, 60, 25]
colors = ['red', 'blue', 'green', 'orange']
```

```python
# Creating vertical barchart
plt.figure(figsize=(8, 6))
plt.bar(subjects, students, color = colors, label = 'Subjects vs Students')
plt.xlabel('Subjects')
plt.ylabel('Students')
plt.title("Subjects vs Students")
plt.show()

# Creating Horizontal barchart 
plt.figure(figsize=(8, 6))
plt.barh(subjects, students, color=colors, label='Subjects vs Students')
plt.xlabel('Students')
plt.ylabel('Subjects')
plt.title('Subjects vs Students')
plt.show()
```

# Pie Chart

```python
labels = ['India', 'Australia', 'NewZealand', 'Srilanka']
sizes = [65, 15,12, 8]
explode = (0.1, 0, 0, 0) 
colors = ['red', 'blue', 'green', 'orange']
```

```python
# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, startangle=90)
plt.title('Pie Chart')
plt.legend()
plt.show()
```

# Multiple Line charts

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 25, 30, 35]
y2 = [5, 10, 20, 25, 30]
y3 = [15, 20, 30, 35, 40]
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='Line 1', marker='o', linestyle='-', color='blue', linewidth=2)
plt.plot(x, y2, label='Line 2', marker='s', linestyle='--', color='red', linewidth=2)
plt.plot(x, y3, label='Line 3', marker='^', linestyle='-.', color='green', linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiline Chart')
plt.legend()
plt.show()
```

# Stacked BarChart

```python
products= ['Product A', 'Product B', 'Product C', 'Product D']
sales1 = [20, 35, 30, 25]
sales2 = [15, 25, 20, 30]
```

```python
# Create stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(products, sales1, label='Group 1', color='green')
plt.bar(products, sales2, bottom=sales1, label='Group 2', color='red')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Company Sales')
plt.legend()
plt.show()
```

```python
colors_c = ['crimson', 'cyan', 'yellow']
origin_counts = df['origin'].value_counts()
```

```python
plt.bar(x = origin_counts.index, height = origin_counts.values, color = colors_c)
plt.xlabel('Total')
plt.ylabel("Continent")
plt.title("Count of cars by origin")
plt.show()
```

# Horizontal barchart

```python
plt.barh(y = origin_counts.index, width = origin_counts.values, color = colors_c)
plt.xlabel('Total')
plt.ylabel("Continent")
plt.title("Count of cars by origin")
plt.show()
```

```python
plt.scatter(df['horsepower'], df['weight'], color = 'b')

plt.xlabel("Horsepower")
plt.ylabel("Weight")
plt.legend()
plt.title("Horsepower VS Weight")
plt.show()
```
