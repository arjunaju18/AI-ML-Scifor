```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('D:\\AI INTERN SCIFOR\\Practice\\revenue.csv')  #importing CSV file
df
```

```python
x = df['Year']
y = df['Revenue']
```

```python
plt.plot(x,y, marker = '*', linestyle = '--', color = 'red', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.show() 
```

```python
import pandas as pd
data = pd.read_csv('D:\\AI INTERN SCIFOR\\Practice\\tips.csv')  #importing CSV file
```

```python
data
```

# Line Plot

```python
plt.figure(figsize=(10, 6))  
plt.plot(data['day'], data['total_bill'], color='blue')
plt.title('Total Bill vs Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['total_bill'], data['tip'], color='blue')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
